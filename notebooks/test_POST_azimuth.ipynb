```python
import pandas as pd
import urllib
import plotly.express as px
```

```python
station = '安丘'
```

```python
params = urllib.parse.urlencode({'stationname': station,
                                'itemname': '第一',
                                'database': 'shandong-official',
                                'startdate': '20240101',
                                'enddate': '20250424'})
url = 'http://localhost:8888/ele_Azimuth?%s' % params
df1 = pd.read_json(url, orient='columns')

df1.index.name = ''
fig = px.scatter(df1, x=df1.index, y=df1.columns,
                trendline_color_override=px.colors.qualitative.Plotly[2],
                labels={'value': '方位角（°）', 'variable': '分量'},
                    )
fig.update_traces(marker_line_width=0.5, marker_line_color='gray',)
fig.update_layout(height=450, width=1200, 
                    font_family='STSong', 
                    font_size=14,
                    template='plotly_white',
                    title=f'{station}长极距优势方位角',
                    )
fig.show()
```
